```python
#N prime. The basis used will be [inf]-[1/N], [inf]-[2/N], ... , [inf]-[floor(N/2)/N]
#N = 4. Basis used [inf]-[1/N]

class Gamma:
    def __init__(self, a, c):      #given numerator and denominator, construct rest of matrix
        """Attributes defined for Gamma:
        a,b,c,d: The matrix coefficients
        num,denom: Numerator and denomnator that generates Gamma by 4.1. Denominator recorded has been divided by N.  
        coeff: +1 or -1, + if c*d >0, - if c*d<0
        x: The numerator in the basis element [inf]-[x/N]
        """
        self.num = int(a)
        self.denom = int(c)
        if c==0:					#case of c = 0 is the infinity case
            self.a = 0
            self.b = 0
            self.c = 0
            self.d = 0				#d will never equal 0 unless its infinity, so this is what we can use to check infinity
            self.coeff = 0
            self.x = 0
        else:
            base = max(N*c,-N*c)
            d = ""
            for x in range(int(-base/2-0.1), int(base/2+0.1)+1):			#in the range... 
                if (a*x) % base == 1:								#find which value satisfies d. (Step is slow, could be optimized later)
                    d=x
            b = (a*d - 1)/(N*c)
            x = max(round(abs(d/c)), 1)								#x is not the unique choice 
            #If N=4, change to =1, else =x. 
            self.x = x
            if c*d > 0:												#second matrix in 4.1
                self.a = int(a-N*b)
                self.b = int(b)
                self.c = int(N*c - N*d)
                self.d = int(d)
                self.coeff = 1
            else:													#first matrix
                self.a = int(a)
                self.b = int(b)
                self.c = int(N*c)
                self.d = int(d)
                self.coeff = -1
    
    def __str__(self):     #for print statements
        return (str(self.a) + " " + str(self.b) + " " + str(self.c) + " " + str(self.d))
    
def getGammas(num, denom):			#returns an array of every gamma used
    a = []							
    gam = Gamma(num, (denom/N))			#constructs the first gamma given the fraction
    a.append(gam)
    return getGammasHelper(gam, a)		#sends to the recursive function
    
def getGammasHelper(prev, gammas):		
    if prev.d == 0:					#if infinity, we are done
        return gammas				
    else:						
        if prev.denom*prev.d > 0:					#second matrix case
            num = prev.num*prev.x - prev.b*N  					#create new fraction
            denom = prev.denom*prev.x*N - prev.d*N
            gam = Gamma(num, (denom/N))					#create new matrix
            gammas.append(gam)
            return getGammasHelper(gam, gammas)			#run it again
        else:
            num = prev.num*prev.x + prev.b*N
            denom = prev.denom*prev.x*N + prev.d*N
            gam = Gamma(num, (denom/N))
            gammas.append(gam)
            return getGammasHelper(gam, gammas)
```

```python
if __name__ == '__main__' and '__file__' not in globals():        
    N = 4
    a = 1
    b = 40
    t = getGammas(a,b)[:-1]  
    decom = "[inf]-" + "[" + str(a) + "/" + str(b) + "] = " + "\n"
    for gamma in t:
        decom += str(gamma.coeff) + "Matrix(" + str(gamma) + ")" + "([inf]-" + "[" + str(gamma.x) + "/" + str(N) + "])" + "\n"
    print(decom)
```
